## <a name="C2.2.1">Mission n° 2.2.1 : Analyse De Concentration, via Une Courbe De Lorenz</a>

<font size="4">
La courbe de Lorenz, du nom de son inventeur, est une représentation graphique permettant de visualiser la distribution d'une variable (actif, patrimoine, revenu, etc.) au sein d'une population. 
</font>

```python
fig_name = 'Courbe De Lorenz de La Variable Price'
fig, ax = plt.subplots()
fig.set_size_inches(12, 9)
depenses = data_all_files[data_all_files['price'] > 0]
dep = depenses['price'].values
n = len(dep)
lorenz = np.cumsum(np.sort(dep)) / dep.sum()
lorenz = np.append([0],lorenz) # La courbe de Lorenz commence à 0

xaxis = np.linspace(0-1/n,1+1/n,len(lorenz)) #Il y a un segment de taille n pour chaque individu, plus 1 segment supplémentaire d'ordonnée 0. Le premier segment commence à 0-1/n, et le dernier termine à 1+1/n.
plt.plot(xaxis,lorenz,drawstyle='steps-post')
plt.plot([0,1], [0,1]) #tracer la bisséctrice
plt.title(fig_name, fontsize=20)
plt.savefig(os.path.join(my_path, fig_name))
plt.show()
```

## <a name="C2.2.2">Mission n° 2.2.2 : Analyse De Concentration, via Un Indice De Gini</a>

<font size="4">
L'indice (ou coefficient) de Gini est un indicateur synthétique permettant de rendre compte du niveau d'inégalité pour une variable et sur une population donnée. Il varie entre 0 (égalité parfaite) et 1 (inégalité extrême).
</font>

```python

AUC = (lorenz.sum() - lorenz[-1]/2 -lorenz[0]/2)/n # Surface sous la courbe de Lorenz. 
#Le premier segment (lorenz[0]) est à moitié en dessous de 0, on le coupe donc en 2, on 
#fait de même pour le dernier segment lorenz[-1] qui est à moitié au dessus de 1.

S = 0.5 - AUC # surface entre la première bissectrice et le courbe de Lorenz
gini = 2*S
gini
```

<p style="font-family: times, serif; font-size:18pt; font-style:normal">
Il est possible d'agréger la variable X en abscisse (axe horizontal) en différentes classes. Cela équivaut à "découper" au couteau le graphique précédent en tranches verticales. On représente ensuite pour chaque tranche une boîte à moustaches calculée à partir de tous les points présents dans la tranche.
</p>


```python
taille_classe = 50 # taille des classes pour la discrétisation

groupes = [] # va recevoir les données agrégées à afficher

# on calcule des tranches allant de 0 au solde maximum par paliers de taille taille_classe
tranches = np.arange(0, max(depenses["price"]), taille_classe)
tranches += taille_classe/2 # on décale les tranches d'une demi taille de classe
indices = np.digitize(depenses["price"], tranches) # associe chaque solde à son numéro de classe
```

```python
fig_name = "Répartition Des Produits Achetés Par Tranches"

for ind, tr in enumerate(tranches): # pour chaque tranche, ind reçoit le numéro de tranche et tr la tranche en question
    montants = depenses.loc[indices==ind,"price"] # sélection des individus de la tranche ind
    if len(montants) > 0:
        g = {
            'valeurs': montants,
            'centre_classe': tr-(taille_classe/2),
            'taille': len(montants),
            'quartiles': [np.percentile(montants,p) for p in [25,50,75]]
        }
        groupes.append(g)

plt.figure(figsize=(10,7))
        
# affichage des boxplots
plt.boxplot([g["valeurs"] for g in groupes],
            positions= [g["centre_classe"] for g in groupes], # abscisses des boxplots
            showfliers= False, # on ne prend pas en compte les outliers
            widths= taille_classe*0.7) # largeur graphique des boxplots

# affichage des effectifs de chaque classe
for g in groupes:
    plt.text(g["centre_classe"],0,"(n={})".format(g["taille"]),horizontalalignment='center',verticalalignment='top')   
plt.title(fig_name)
plt.xlabel("Tranche De Prix (n = Nombre D'achats)", fontsize=12)
plt.ylabel("Prix", fontsize=12)
plt.savefig(os.path.join(my_path, fig_name))
plt.show()
```


# <a name="C4.1">Test d'indépendance avec le Chi-2 : Catégorie Et Âge (Test du chi2 ou Khi2 ou χ²)</a>

https://support.minitab.com/fr-fr/minitab/18/help-and-how-to/statistics/tables/how-to/chi-square-test-for-association/interpret-the-results/all-statistics/

<p style="font-family: times, serif; font-size:18pt; font-style:normal">
   Ici on va utiliser la méthode du cours pour avoir une heatmap
</p>

<p style="font-family: times, serif; font-size:18pt; font-style:normal">
   Autre méthode 1
</p>

```python
X = "categ"
Y = "sex"

cont = data_all_files[[X,Y]].pivot_table(index=X,columns=Y,aggfunc=len)#,margins=True,margins_name="Total")

tx = pd.DataFrame(data_all_files[X].value_counts())  #cont.loc[:,["Total"]]
ty = pd.DataFrame(data_all_files[Y].value_counts() ) #cont.loc[:,["Total"]]

n = len(data_all_files)
ty.columns=['m']
tx.columns=['m']

indep = ty.dot((tx.T))/n
##########################################################################################

c = cont.fillna(0) # On remplace les valeurs nulles par 0
measure = (c.T-indep)**2/indep
xi_n = measure.sum().sum()
table = measure/xi_n
##########################################################################################

heatmap = sns.heatmap(table,annot=True, cmap='plasma')
heatmap.set_title('Tableau De Contingence Coloré', fontdict={'fontsize':18}) 
plt.xlabel('categ', fontsize=18)
plt.ylabel('Sex', fontsize=18)
plt.show()
```

```python
# Tableau de contingence entre les variables 'sex' et 'catégorie'
df = data_all_files
cont_table = pd.crosstab(df['categ'], df['sex'])
cont_table
```

```python
fig_name = 'Heatmap de sexe et la catégorie'
sns.heatmap(cont_table)#, annot=True)
plt.title('Heatmap entre le sexe et la catégorie')
plt.savefig(os.path.join(my_path, fig_name))
plt.ylabel('categ', fontsize=18)
plt.xlabel('Sex', fontsize=18)
plt.show()
```

<p style="font-family: times, serif; font-size:18pt; font-style:normal">
Test Chi2 :
</p>

```python
# Hypothèses :
# Ho : La catégorie est indépendante du sex
# Hα : La catégorie est dépendante ou est associée au sex
```

```python
# Test du Chi2
α = 0.05
cont_table = df#.sample(5000)
cont_table = pd.crosstab(cont_table['categ'], df['sex'])
chi2float, pfloat, dofint, f = st.chi2_contingency(cont_table)#, correction = True)

# Ce que renvoi la fonction scipy.stats.chi2_contingency(dt)
'''
Returns
        chi2float
        The test statistic.

        pfloat
        The p-value of the test

        dofint
        Degrees of freedom

expectedndarray, same shape as observed
The expected frequencies, based on the marginal sums of the table.
'''
print(bcolors.GREEN + '\n\nTest d\'indépendance avec le Chi-2 :' + bcolors.RESET+'\n')
print(bcolors.RED + 'The test statistic = %.2f'% (chi2float) + bcolors.RESET+'\n')
print(bcolors.RED + 'The p-value of the test = %e' % (pfloat) + bcolors.RESET+'\n')
print(bcolors.RED + 'Degrees of freedom = %.2f' % (dofint) + bcolors.RESET+'\n')


if pfloat > α: #(On choisit un risque standard de α = 5% ou 0.05 )
    print('Conclusion : Impossible de conclure que les variables sont associées (Ne pas rejeter H0)')
    print('             Possibilité d\'indépendance des variable categ et sex')
else:
    print('Conclusion : On rejete H0, on garde Hα, il existe une association statistiquement') 
    print('             significative entre les variables sex & categ ')
    

```
